## REUSABLE CODE IMPLEMENTATIONS
### Functional Definitions
```python
def AND(a, b): return a * b
def OR(a, b): return 1 - ((1 - a) * (1 - b))
def NOT(a): return 1 - a
def NAND(a, b): return 1 - (a * b)
def XOR(a, b): return OR(a, b) * NAND(a, b)
```

### Truth Table Verification Script
```python
bool_dict_1 = {0:0, 1:1}
bool_dict_2 = {1:0, 0:1}

for a,b in bool_dict_1.items():
    AND = a * b
    NAND = 1 - (a * b)
    OR = 1 - ((1-a) * (1-b))
    XOR = (1 - ((1-a) * (1-b))) * (1 - (a * b))
    print(f"a={a}, b={b} -> AND={AND}, NAND={NAND}, OR={OR}, XOR={XOR}")

for a,b in bool_dict_2.items():
    AND = a * b
    NAND = 1 - (a * b)
    OR = 1 - ((1-a) * (1-b))
    XOR = (1 - ((1-a) * (1-b))) * (1 - (a * b))
    print(f"a={a}, b={b} -> AND={AND}, NAND={NAND}, OR={OR}, XOR={XOR}")
```
